```python
import matplotlib.pyplot as plt
import numpy as np
```

## 25 steps

**High order strength**

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

width = 0.15

cycle_time_color = 'lightblue'
std_dev_color = 'darksalmon'

scenarios = [
    {'name': 'Scenario A', 'cycle_time': 598.98, 'std_dev': 37.08},
    {'name': 'Scenario B.1', 'cycle_time': 572.04, 'std_dev': 14.52},
    {'name': 'Scenario B.2', 'cycle_time': 452.7, 'std_dev': 69.84},
    {'name': 'Scenario C', 'cycle_time': 575.05, 'std_dev': 92.46},
    {'name': 'Scenario D.1', 'cycle_time': 701.35, 'std_dev': 579.97},
    {'name': 'Scenario D.2', 'cycle_time': 651.85, 'std_dev': 184.63},
    {'name': 'Scenario D.3', 'cycle_time': 548.85, 'std_dev': 349.86},
    {'name': 'Scenario E', 'cycle_time': 448.65, 'std_dev': 31.55},
]


x_values = np.arange(len(scenarios))
cycle_time_values = [s['cycle_time'] for s in scenarios]
std_devs = [s['std_dev'] for s in scenarios]
scenario_names = [s['name'] for s in scenarios]

# Plot cycle time on left y-axis
bars1 = ax1.bar([x - width/2 for x in x_values], cycle_time_values, width, color=cycle_time_color, 
                label='Cycle Time')
ax1.set_xlabel('Scenario')
ax1.set_ylabel('Cycle Time')
ax1.set_xticks(x_values)
ax1.set_xticklabels(scenario_names)

ax2 = ax1.twinx()

# Plot standard deviation on right y-axis
bars2 = ax2.bar([x + width/2 for x in x_values], std_devs, width, color=std_dev_color, 
                label='Smoothness Score')
ax2.set_ylabel('Smoothness Score')

plt.title('Trade-off between Cycle Time and Smoothness Score for different scenarios (high order strength)', fontsize=12)
legend_elements = [
    plt.Rectangle((0,0), 1, 1, color=cycle_time_color, label='Cycle Time'),
    plt.Rectangle((0,0), 1, 1, color=std_dev_color, label='Smoothness Score'),
]
ax1.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=3)

plt.tight_layout()
plt.show()
```
